```python
import mysql.connector
from mysql.connector import Error
import boto3

# RDS details
rds_host = "youtubedb.c96m2ysgag7t.eu-north-1.rds.amazonaws.com"
db_name = "youtubedb"
username = "admin"
password = "nambukeerthi"
table_name = "employees"  # Replace with your table name

try:
    # Connect to MySQL server
    conn = mysql.connector.connect(
        host=rds_host,
        user=username,
        password=password,
        database=db_name
    )
    if conn.is_connected():
        print("Connected to MySQL database!")
# table creation 
        create_table_query = """
            CREATE TABLE IF NOT EXISTS employees (
                id INT AUTO_INCREMENT PRIMARY KEY,
                name VARCHAR(100) NOT NULL,
                age INT,
                department VARCHAR(50),
                salary DECIMAL(10, 2)) """
        cursor = conn.cursor()
        cursor.execute(create_table_query)
        print("Table 'employees' created successfully.")
#insert values 
        insert_query = """
            INSERT INTO employees (name, age, department, salary)
            VALUES (%s, %s, %s, %s)
        """
        data_to_insert = ("John Doe", 30, "HR", 5000.00)


        cursor.execute(insert_query, data_to_insert)
        conn.commit()
        print("Data inserted successfully.")

        # Querying data from the table
        select_query = "SELECT * FROM employees"
        cursor.execute(select_query)

        print("\nFetching data from 'employees' table:")
        rows = cursor.fetchall()
        for row in rows:
            print(f"ID: {row[0]}, Name: {row[1]}, Age: {row[2]}, Department: {row[3]}, Salary: {row[4]}")
        

        drop_table_query = f"DROP TABLE IF EXISTS {table_name}"

        cursor = conn.cursor()
        cursor.execute(drop_table_query)
        conn.commit()
        print(f"Table '{table_name}' dropped successfully.")

        # Closing cursor and connection
        cursor.close()

        # Check if the table exists
        cursor = conn.cursor()
        cursor.execute(f"SHOW TABLES LIKE '{table_name}'")
        result = cursor.fetchone()

        if result:
            print(f"Table '{table_name}' exists.")
        else:
            print(f"Table '{table_name}' does not exist.")

        # Close cursor and connection
        cursor.close()

except Error as e:
    print("Error connecting to MySQL database:", e)

finally:
    # Close database connection
    if conn is not None and conn.is_connected():
        conn.close()
        print("MySQL connection closed.")
```
